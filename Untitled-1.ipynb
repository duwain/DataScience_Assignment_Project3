```python
from sqlalchemy import create_engine
# Create connection string using credentials following this format
# connection = "dialect+driver://username:password@host:port/database"
username = "root"
password = "root" # (or whatever password you chose during mysql installation)
db_name = "sakila"
connection = f"mysql+pymysql://{username}:{password}@localhost/{db_name}"

```

```python
engine = create_engine(connection)
```

```python
engine
```

```python
q = """SELECT * FROM customer;"""
pd.read_sql(q, engine)
```

```python
query = """
SELECT address.city_id AS city_id, customer.first_name AS first_name, customer.last_name AS last_name, customer.email, address.address
FROM customer
JOIN address ON customer.address_id = address.address_id
WHERE address.city_id = 312;
"""
pd.read_sql(query, engine)
```

```python
query = """
    SELECT * FROM film
    JOIN film_category ON film.film_id = film_category.film_id
    JOIN category ON film_category.category_id = category.category_id
    WHERE category.name = 'Comedy';
    """
pd.read_sql(query, engine)
```

```python
query = """
SELECT actor.last_name AS last_name , film.title, film.release_year
FROM film
JOIN actor ON actor.actor_id = actor.actor_id
WHERE actor.last_name = 'Lollobrigida';
"""
pd.read_sql(query, engine)
```

```python
query=""" 
SELECT film.title, actor.first_name, actor.last_name
FROM film
JOIN actor ON actor.actor_id = actor.actor_id
WHERE film.title ='Bingo Talented';
"""
pd.read_sql(query, engine)
```

```python
query= """
SELECT customer_id, amount
FROM payment
WHERE amount > 2 * (SELECT AVG(amount) FROM payment);
"""
pd.read_sql(query, engine)
```

```python
query="""
SELECT c.first_name, c.last_name, COUNT(p.customer_id) AS num_payments
FROM customer AS c
JOIN payment AS p ON c.customer_id = p.customer_id
GROUP BY c.customer_id
ORDER BY num_payments DESC
LIMIT 5;
"""
pd.read_sql(query, engine)
```
